```python
#Data update script for list of castles in Japan

import pandas as pd
import numpy as np
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_National_Treasures_of_Japan_(castles)"
```

```python
tables = pd.read_html(url)
len(tables)
```

```python
#tables are a list of dataframes 
#When you find the table you want, call it df 
df = tables[1]
df
```

```python
df['Castle'].value_counts()
```

```python
#to_csv will save the dastaframe as a csv file
#and index=False will remove the index column
df.to_csv("castles.csv", index=False)
```
